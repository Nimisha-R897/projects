```python
#training--------
#1.Load the DATA (excel sheet)
import pandas as pd
file=pd.read_csv("/content/irisexcel.csv")
x=file[["sepal_length",  "sepal_width",  "petal_length",  "petal_width" ]]
y=file["species"]

#Cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)

#2.Load the MODEL(perceptron)
from sklearn.linear_model import Perceptron
a=Perceptron()

#3.Fit DATA into MODEL
a.fit(xtrain,ytrain)

#validation
op=a.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("accuracy=",score*100)

#Testing--------predict
op2=a.predict([[1.5,2.5,3.5,4.5]])
print(op2)
```
